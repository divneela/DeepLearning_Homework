```python
import fitz
import requests
from bs4 import BeautifulSoup
import openai
```

```python
# Function to interact with GPT-3.5 Turbo
def get_completion_from_messages(messages, model="gpt-3.5-turbo", temperature=0):
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=temperature, # this is the degree of randomness of the model's output
    )
    return response.choices[0].message["content"]
```

```python
# Initial message setup
messages = [
    {"role": "system", "content": "You are an assistant capable of solving complex math problems, summarizing texts, and answering specific content questions.if the answer is not known , say i dont know"},
]
```

# a. Ask the bot to solve one complex math problem.

```python
# Math problem interaction
add_message = lambda content: messages.append({"role": "user", "content": content})
add_message("Can you help me solve the integral of 3x^2 from 0 to 1?")
print(get_completion_from_messages(messages))
```

```python
add_message( "Can you find the eigenvalues and eigenvectors of the matrix [[4, 1], [2, 3]]?")
print(get_completion_from_messages(messages))
```

#  b. Give a PDF and website document; ask the bot to rewrite and answer questions on the given PDF and website.

```python



# Function to extract text from PDF
def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text()
    return text

# Function to scrape text from a website
def scrape_text_from_url(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    return ' '.join([p.text for p in soup.find_all('p')])  # Extracting text from paragraph tags

```

```python

# PDF text processing
pdf_path = '/content/drive/MyDrive/DL_HW11/story.pdf'
pdf_text = extract_text_from_pdf(pdf_path)
add_message(f"what is the 1st story heading from a PDF document:\n\n{pdf_text}")
print("PDF Summary:", get_completion_from_messages(messages))

```

```python


add_message(f"what is the moral of 2nd story from a PDF document:\n\n{pdf_text}")
print("PDF Summary:", get_completion_from_messages(messages))

```

```python

add_message(f"summarize the 2nd story from a PDF document:\n\n{pdf_text}")
print("PDF Summary:", get_completion_from_messages(messages))

```

```python
# Website content interaction
website_url = 'https://webscraper.io/test-sites/e-commerce/allinone/computers'
website_text = scrape_text_from_url(website_url)
add_message(f"what are the products mentioned in the website?:\n\n{website_text}")
print("Website Products:", get_completion_from_messages(messages))
```
